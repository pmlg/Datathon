```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm, colors
import cv2
import os
%matplotlib inline
```

### Based on 3.tif, blue channel was extracted for nuclei only. 

```python
img = cv2.imread('3_b.png')
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
ret, thresh = cv2.threshold(gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)

plt.imshow(cv2.cvtColor(thresh, cv2.COLOR_BGR2RGB))
```

```python
# noise removal
kernel = np.ones((3,3),np.uint8)
opening = cv2.morphologyEx(thresh,cv2.MORPH_OPEN,kernel, iterations = 2)

# sure background area
sure_bg = cv2.dilate(opening,kernel,iterations=3)
```

```python
# Finding sure foreground area
dist_transform = cv2.distanceTransform(opening,cv2.DIST_L2,5)  # option 0, 3 & 5
ret, sure_fg = cv2.threshold(dist_transform,0.7*dist_transform.max(),255,0)

# Finding unknown region
sure_fg = np.uint8(sure_fg)
unknown = cv2.subtract(sure_bg,sure_fg)
plt.imshow(cv2.cvtColor(sure_fg, cv2.COLOR_BGR2RGB))
plt.show()

plt.imshow(cv2.cvtColor(unknown, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
# Marker labelling
ret, markers = cv2.connectedComponents(sure_fg)

# Add one to all labels so that sure background is not 0, but 1
markers = markers+1

# Now, mark the region of unknown with zero
markers[unknown==255] = 0
```

```python
plt.imshow(markers, cmap=cm.jet,alpha=0.75)
plt.show()
```

```python
markers = cv2.watershed(img,markers)
img[markers == -1] = [255,0,0]

plt.figure(figsize=(10,10))
plt.title('Fig 1. Watershed Mask')
plt.imshow(markers, cmap=cm.jet,alpha=0.75)
plt.show()

plt.figure(figsize=(10,10))
plt.title('Fig 2. Watershed Outline')
plt.imshow(img, cmap=cm.jet,alpha=0.75)
plt.show()

plt.figure(figsize=(10,10))
img = cv2.imread('3.tif')
plt.title('Fig 3. Original image - 3.tiff')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()

plt.figure(figsize=(10,10))
img = cv2.imread('3_b.png')
plt.title('Fig 4. 3.tif blue channel only')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

### Result: In 3.tif image, majority of nuclei of the right hand side were not easily visible to observer.  After applied watershed aligorithum (Fig 2),  those nuclei became visible (red/brown colour).  However, the aligorithum could not be detected its outline.
